```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
```

```python
def cost(x):
    tar = 4
    return (x-tar) ** 2
```

```python
def num_slope(b): 
    h = 0.0001
    return (cost(b+h) - cost(b))/h
```

```python
cost(3.98)
```

```python
num_slope(3.98)
```

```python
def slope_calc(b):
    return 2 * (b-4)
```

```python
def wags_pats(x, w, b):
    arr = np.array([0.,0.,0.])
    for i in range(3):
        print(x)
        arr[i] = w*x[i] + b
        print(arr[i])
    return arr
    
```

```python
w = 0.25
b = 0.68
pats = np.array([1,2,4])
wp_arr = wags_pats(pats, w, b)
wp_arr2 = wags_pats(pats, 0.44, 0.56)
plt.plot(pats, wp_arr)
plt.plot(pats, wp_arr2)


```
